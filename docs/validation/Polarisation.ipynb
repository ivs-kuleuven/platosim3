```python
from simfile import SimFile
from simulation import Simulation
from validation import switchOffAllEffects
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
sim = Simulation("Polarisation")
switchOffAllEffects(sim)
sim.outputDir = os.environ["PLATO_WORKDIR"]

# One full-frame exposure

sim["ObservingParameters/NumExposures"] = 1
sim["SubField/NumRows"] = 4510
sim["SubField/NumColumns"] = 4510
```

```python
# Enable polarisation

sim["CCD/IncludePolarization"] = "yes"
imageWithPolarization = sim.run(removeOutputFile = True).getImage(0)

# Disable polarisation

sim["CCD/IncludePolarization"] = "no"
imageWithoutPolarization = sim.run(removeOutputFile = True).getImage(0)
```

```python
quotient = imageWithPolarization / imageWithoutPolarization
```

```python
polarization = sim["CCD/Polarization"]
```

```python
minRatio = np.min(quotient)
maxRatio = np.max(quotient)
```

```python
print("Polarisation efficiency as estimated in PlatoSim:", minRatio, " -", maxRatio)
print("To compare with the expected value from the configuration file", polarization)
```
