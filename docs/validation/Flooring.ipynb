```python
from simfile import SimFile
from simulation import Simulation
from validation import switchOffAllEffects
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
sim = Simulation("Flooring")
switchOffAllEffects(sim)
sim.outputDir = os.environ["PLATO_WORKDIR"]
```

```python
# One full-frame exposure

sim["ObservingParameters/NumExposures"] = 1
sim["SubField/NumRows"] = 4510
sim["SubField/NumColumns"] = 4510

# Enable quantisation (flooring only)

sim["CCD/IncludeQuantisation"] = "yes"
sim["FEE/Gain/RefValueLeft"] = 1
sim["FEE/Gain/RefValueRight"] = 1
sim["CCD/Gain/RefValueLeft"] = 1
sim["CCD/Gain/RefValueRight"] = 1
sim["FEE/ElectronicOffset/RefValue"] = 0

sim["CCD/IncludeDigitalSaturation"] = "yes"
saturationLimit = sim["CCD/DigitalSaturation"]

output = sim.run(removeOutputFile = True)
imageWithFlooring = output.getImage(0)

# Disable quantisation

sim["CCD/IncludeQuantisation"] = "no"
output = sim.run(removeOutputFile = True)
imageWithoutFlooring = output.getImage(0)
```

```python
difference = imageWithoutFlooring[imageWithoutFlooring <= saturationLimit] - imageWithFlooring[imageWithoutFlooring <= saturationLimit]
```

```python
print(np.min(difference))
print(np.max(difference))
```
