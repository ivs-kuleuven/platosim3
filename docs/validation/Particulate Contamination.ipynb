```python
from simfile import SimFile
from simulation import Simulation
from validation import switchOffAllEffects
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
sim = Simulation("ParticulateContamination")
switchOffAllEffects(sim)
sim.outputDir = os.environ["PLATO_WORKDIR"]

# One full-frame exposure

sim["ObservingParameters/NumExposures"] = 1
sim["SubField/NumRows"] = 4510
sim["SubField/NumColumns"] = 4510
```

```python
sim["CCD/IncludeParticulateContamination"] = "yes"
imageWithParticulateContamination = sim.run(removeOutputFile = True).getImage(0)

sim["CCD/IncludeParticulateContamination"] = "no"
imageWithoutParticulateContamination = sim.run(removeOutputFile = True).getImage(0)
```

```python
quotient = imageWithParticulateContamination / imageWithoutParticulateContamination
```

```python
particulateContamination = sim["CCD/Contamination/ParticulateContaminationEfficiency"]
```

```python
minRatio = np.min(quotient)
maxRatio = np.max(quotient)
```

```python
print("Polarisation efficiency as estimated in PlatoSim:", minRatio, " -", maxRatio)
print("To compare with the expected value from the configuration file", particulateContamination)
```
