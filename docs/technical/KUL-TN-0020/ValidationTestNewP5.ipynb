## Validation of fixed seed 1234

```python
# Load all data for a single star
idir = "/lhome/nicholas/software/workdir/kul20/star50"
lc1 = LightCurve(f"{idir}/test1_000000050_Ncam1.1_Q23.hdf5")
lc2 = LightCurve(f"{idir}/test2_000000050_Ncam1.1_Q23.hdf5")
lc3 = LightCurve(f"{idir}/test1_000000050_Ncam1.1_Q23.ftr")
lc4 = LightCurve(f"{idir}/test2_000000050_Ncam1.1_Q23.ftr")
df1 = lc1.data()
df2 = lc2.data()
df3 = lc3.data()
df4 = lc4.data()
t0 = df1.time[0]
```

```python
plt.figure(figsize=(9,7))
plt.plot(df1.time - t0, df1.flux, 'ro', label='PlatoSim 1')
plt.plot(df2.time - t0, df2.flux, 'k.', label='PlatoSim 2')
plt.plot(df3.time - t0, df3.flux, 'bo', label='Pipeline 1')
plt.plot(df4.time - t0, df4.flux, 'k.', label='Pipeline 2')
plt.plot(df3.time - t0, df3.flux/df3.flux_cor, 'go', label='Pipeline/cor 1')
plt.plot(df4.time - t0, df4.flux/df4.flux_cor, 'k.', label='Pipeline/cor 2')
plt.legend()
plt.xlabel('Time [s]')
plt.ylabel('Flux [e/s]')
plt.tight_layout()
```
