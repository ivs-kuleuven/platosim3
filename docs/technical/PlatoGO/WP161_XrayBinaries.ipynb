````python
# Query sources from Gaia DR3 database
df = pd.DataFrame()
for n in hmxb:
    try:
        dx = sq.simbadQuery(n, radius=15, maglim=21)
        dx['name'] = n
    except:
        print(n)
        pass
    else:
        df = pd.concat([df, dx.iloc[0].to_frame().T])
````

````python
cols = list(df)
cols.insert(0, cols.pop(cols.index('name')))
df = df.loc[:, cols]
df = df.drop(columns=['dis'])
df
````
