## LOPS2PICtarget2.1.0.1-t-fg-c-scv

This catalogue is the main target table, composed by the merge of four different target lists:

- `tPIC`: core science stars (P1, P2, P4, P5) and stars with confirmed/candidate planet(s)
- `fgPIC`: Fine Guidance System (FGS) stars used to guarantee the attitude of the spacecraft
- `cPIC`: instrument calibration stars
- `scvPIC`: calibration and validation stars for both stellar and exoplanets science

Load target catalogue into data frame:

```python
table = Table.read(path / 'LOPS2PICtarget2.1.0.1-t-fg-c-scv.fits', format='fits')
df = table.to_pandas()
df.head()
```

Check the column names:

```python
df.columns
```

Define a new dataframe and copy over the columns we need:

```python
dt = pd.DataFrame()
dt['PIC'] = df.PICid
dt['gaiaDR3'] = df.StarName
dt['ra'] = df.RAdeg
dt['dec'] = df.DEdeg
dt['pmRA'] = df.pmRA
dt['pmDec'] = df.pmDE
dt['Pmag'] = df.PlatoMagNCAM
dt['PBmag'] = df.PlatoMagFCAMb
dt['PRmag'] = df.PlatoMagFCAMr
dt['Teff'] = df.Teff
dt['R'] = df.Radius
dt['M'] = df.Mass
dt['case'] = df.caseFlag
dt['source'] = df.PICmainSourceFlagBOL
dt['tPIC'] = df.tPICsourceFlagNCAM_BOL
dt['fgPIC'] = df.tPICsourceFlagNCAM_BOL
dt['cPIC'] = df.tPICsourceFlagNCAM_BOL
dt['scvPIC'] = df.tPICsourceFlagNCAM_BOL
dt['ncams'] = df.BOLnCameraObsNCAM_T
# dt[''] = df.
```

Change bitmask values of source flag defining the sub-PICs:

```python
dt['source'].replace(0, np.nan, inplace=True)
dt['source'].replace(1, 'Prime', inplace=True)
dt['source'].replace(2, 'Proprietary', inplace=True)
dt['source'].replace(4, 'tPIC', inplace=True)
dt['source'].replace(8, 'fgPIC', inplace=True)
dt['source'].replace(16, 'cPIC', inplace=True)
dt['source'].replace(32, 'scvPIC', inplace=True)
```

Change bitmask values of source flag defining the target-PIC (tPIC): 

```python
dt['tPIC'].replace(0, np.nan, inplace=True)
dt['tPIC'].replace(1, 'P1', inplace=True)
dt['tPIC'].replace(2, 'P2', inplace=True)
dt['tPIC'].replace(4, 'P5', inplace=True)
dt['tPIC'].replace(8, 'P4', inplace=True)
dt['tPIC'].replace(16, 'Planet', inplace=True)
dt['tPIC'].replace(32, 'Prime', inplace=True)
dt['tPIC'].replace(64, 'Proprietary', inplace=True)
```

Change bitmask values of source flag defining the fine-guidance-PIC (fgPIC): 

```python
dt['fgPIC'].replace(0, 'normal', inplace=True)
dt['fgPIC'].replace(1, 'blue', inplace=True)
dt['fgPIC'].replace(2, 'red', inplace=True)
```

Change bitmask values of source flag defining the calibration-PIC (cPIC): 

```python
dt['cPIC'].replace(0, 'Reset', inplace=True)
dt['cPIC'].replace(1, '', inplace=True)
dt['cPIC'].replace(2, '', inplace=True)
dt['cPIC'].replace(4, '', inplace=True)
dt['cPIC'].replace(8, '', inplace=True)
dt['cPIC'].replace(16, '', inplace=True)
dt['cPIC'].replace(32, '', inplace=True)
dt['cPIC'].replace(64, '', inplace=True)
dt['cPIC'].replace(128, '', inplace=True)
dt['cPIC'].replace(256, '', inplace=True)
dt['cPIC'].replace(512, '', inplace=True)
```
