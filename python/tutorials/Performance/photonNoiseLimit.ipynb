# Photon noise limit

```python
import numpy as np
import matplotlib.pyplot as plt
from platosim.utilities import passbandConversionV2P, NSRphotonNoiseLimit

# Handle warnings
import warnings
warnings.filterwarnings('ignore')

# Change matplotlib defautls
plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15) 
plt.rcParams.update({'font.size': 15})
plt.rcParams['text.usetex'] = True
```

## User defined parameters

```python
Teff = 6000
V = np.arange(0.0, 13.5, 0.05)
```

```python
# Coversions from Marchiori et al. (2019):
P  = passbandConversionV2P(V, Teff)
```

## Photon noise limit vs. N-CAM visibility

```python
plt.figure(figsize=(8,7))

plt.axhline(100, ls='-',  c='k', label=r'Gran+Puls Sun', lw=1)
plt.axhline(84,  ls='--', c='k', label=r'$\delta_{tra}$ Earth-Sun', lw=1)
plt.axhline(28,  ls='-.', c='k', label=r'$\delta_{tra}$ Mars-Sun', lw=1)
plt.axhline(5.1, ls=':',  c='k', label=r'$\delta_{tra}$ Exomoon')

plt.plot(P, NSRphotonNoiseLimit(P, 6,  1, 25.), '--', color='orange',  label='06 Cam, 1 cyc')
plt.plot(P, NSRphotonNoiseLimit(P, 12, 1, 25.), '--', color='green',   label='12 Cam, 1 cyc')
plt.plot(P, NSRphotonNoiseLimit(P, 18, 1, 25.), '--', color='blue',    label='18 Cam, 1 cyc')
plt.plot(P, NSRphotonNoiseLimit(P, 24, 1, 25.), '--', color='magenta', label='24 Cam, 1 cyc')

plt.plot(P, NSRphotonNoiseLimit(P, 6),  '-', color='orange',  label='06 Cam, 1 h')
plt.plot(P, NSRphotonNoiseLimit(P, 12), '-', color='green',   label='12 Cam, 1 h')
plt.plot(P, NSRphotonNoiseLimit(P, 18), '-', color='blue',    label='18 Cam, 1 h')
plt.plot(P, NSRphotonNoiseLimit(P, 24), '-', color='magenta', label='24 Cam, 1 h')

plt.plot(P, NSRphotonNoiseLimit(P, 6,  2.), ':', color='orange',  label='06 Cam, 1 h, 2 tra')
plt.plot(P, NSRphotonNoiseLimit(P, 12, 2.), ':', color='green',   label='12 Cam, 1 h, 2 tra')
plt.plot(P, NSRphotonNoiseLimit(P, 18, 2.), ':', color='blue',    label='18 Cam, 1 h, 2 tra')
plt.plot(P, NSRphotonNoiseLimit(P, 24, 2.), ':', color='magenta', label='24 Cam, 1 h, 2 tra')

plt.ylabel('NSR [ppm]')
plt.xlabel(r'PLATO passband, $P$')
plt.legend(ncol=2)
plt.xlim(V[0], P[-1])
plt.ylim(0.1, 10000)
plt.yscale('log')
plt.tight_layout()
plt.show()
```
